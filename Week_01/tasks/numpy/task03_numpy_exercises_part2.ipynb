#### Task 4: Python vs NumPy

Use pure Python and `numpy` to create a Python list/`numpy`:

- array of $10$ zeros;
- array of $10$ ones;
- array of $10$ fives;
- array of $10$ zeroes, the fifth value of which is $1$;
- array of all the integers from $10$ to $50$;
- array of all the even integers from $10$ to $50$;
- array that is the reversal of the list `[1, 2, 3]` (save these values in an object and use both methods to reverse that object);
- a 3x3 matrix with values ranging from $0$ to $8$;
- a 3x3 identity matrix;
- a 5x5 matrix with values 1,2,3,4 just below the diagonal.

````python
zeroes_array = [0 for i in range(0, 10)]
zeroes_array_np = np.zeros(10)
print(
    f"array of 10 zeros \n Python: {zeroes_array} \n Numpy: {zeroes_array_np} \n")

ones_array = [1 for i in range(0, 10)]
ones_array_np = np.ones(10)
print(f"array of 10 ones \n Python: {ones_array} \n Numpy: {ones_array_np} \n")

fives_array = [5 for i in range(0, 10)]
fives_array_np = np.array(fives_array, dtype=float)
print(f"array of 10 fives \n Python: {fives_array} \n Numpy: {fives_array_np} \n")

mixed_array = []
for num in range(0, 10):
    if (num == 4):
        mixed_array.append(1)
    else:
        mixed_array.append(0)
mixed_array_np = np.eye(10)[4]
print(
    f"array of 10 zeroes, the fifth value of which is 1 \n Python: {mixed_array} \n Numpy: {mixed_array_np}\n"
)

nums_array = []
for num in range(10, 50):
    nums_array.append(num)
nums_array_np = np.arange(start=10, stop=51)
print(
    f"array of all the integers from 10 to 50 \n Python: {nums_array} \n Numpy: {nums_array_np}\n"
)

nums_even_array = []
for num in range(10, 51):
    if (num % 2 == 0):
        nums_even_array.append(num)
nums_even_array_np = np.arange(start=10, stop=51, step=2)
print(
    f"array of all the even integers from 10 to 50 \n Python: {nums_even_array} \n Numpy: {nums_even_array_np}\n"
)

array_to_reverse = [1, 2, 3]
reversed_array = list(reversed(array_to_reverse))
reversed_array_np = np.flip(array_to_reverse)
print(
    f"array that is the reversal of the list `[1, 2, 3]` \n Python: {reversed_array} \n Numpy: {reversed_array_np}\n"
)
array_to_reverse = [1, 2, 3]
reversed_array = list(reversed(array_to_reverse))
reversed_array_np = np.flip(array_to_reverse)
print(
    f"array that is the reversal of the list `[1, 2, 3]` \n Python: {reversed_array} \n Numpy: {reversed_array_np}\n"
)

matrix_array = []
for i in range(0, 3):
    row = []
    for j in range(0, 3):
        row.append(i * 3 + j)
    matrix_array.append(row)
matrix_array_np = np.arange(9).reshape(3, 3)
print(
    f"a 3x3 matrix with values ranging from 0 to 8 \n Python: {matrix_array} \n Numpy: {matrix_array_np}\n"
)
id_matrix_array = []
for i in range(0, 3):
    row = []
    for j in range(0, 3):
        row.append(1 if i == j else 0)
    id_matrix_array.append(row)
id_matrix_array_np = np.identity(3)
print(
    f"a 3x3 identity matrix \n Python: {id_matrix_array} \n Numpy: {id_matrix_array_np}\n"
)
diagonal_matrix_array = []
for i in range(0, 5):
    row = []
    for j in range(0, 5):
        row.append(i if i == j + 1 else 0)
    diagonal_matrix_array.append(row)
diagonal_matrix_array_np = np.diag([1,2, 3, 4], k=-1)
print(
    f"a 5x5 matrix with values 1,2,3,4 just below the diagonal \n Python: {diagonal_matrix_array} \n Numpy: {diagonal_matrix_array_np}\n"
)

````
